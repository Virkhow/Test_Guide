```python


import matplotlib.pyplot as plt
import numpy as np
import math as m

xn = np.arange(-2, 2, 0.1)
y = xn**3 -2*xn**2 -11*xn +12

plt.figure(1)
plt.plot(xn,y, alpha=0.7)
plt.scatter(1,0, color="red", s=40, alpha=1)
plt.grid(alpha=0.4)
plt.show()
# вижу что уравнение равно 0 в промежутке икса от 0.5 до 1.5


from scipy.optimize import bisect

def my_fun(x):
    return x**3 -2*x**2 -11*x +12

x = bisect(my_fun, 0.5, 1.5, xtol=2e-12)

print('Корень уравнения= ', x)



def f(x):
  return x**3 - 2*x**2 - 11*x + 12

def find(a, b, e):
  inter = 0
  while abs(b - a) > e:
    c = (a+b)/2
    if f(a) * f(c) <= 0:
      b = c
    else:
      a = c

    inter += 1
  return (a+b)/2, inter

e = 1e-15

x2, inttt = find(0.5, 1.5, e)

print(f"Корень по интервалам = {x2:.2}, интераций: {inttt}")
print(f"Разница {x2 - x}")



```
